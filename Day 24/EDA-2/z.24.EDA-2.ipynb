```python
# ISOLATION FOREST
```

```python
from sklearn.ensemble import IsolationForest
```

```python
data4 =pd.read_csv("D:/DATA SCIENCE/1. Data Scientist/Day 24- EDA (part 2)/iris.csv",index_col=0)
data_encoded=pd.get_dummies(data4)
```

```python
data4.head()
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
# randomly select subset OF DATA, .01 = 1% value assigned for their outliers(domain knowledge based)
```

```python
# predictions
y_pred_outliers = clf.predict(data_encoded)
```

```python
#-1 for outliers and 1 for inliers.
y_pred_outliers
```

```python
# 1 = not otlier , -1 = outlier
```

```python
data_encoded
```

```python
## Let us add a new data point which is outlier
# adding one row that having high values as compared all dataset, idea to show it shows outlier or not
data_encoded.loc[150]=[20,40,30,50,1,0,0]
```

```python
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
# predictions
y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers
# last point is -1 see 
```

```python
data_encoded['scores'] = clf.decision_function(data_encoded)
```

```python
data_encoded['anomaly'] = clf.predict(data_encoded.iloc[:,0:7])
# wherever you see negative value that is otlier/anomoly
```

```python
data_encoded
```

```python
#Print the outlier data points
# let us filter anomoly/outlier
data_encoded[data_encoded['anomaly']==-1]
```
